```python
sen_model = SentenceTransformer('joeddav/xlm-roberta-large-xnli')
```

```python
sentences = []
y = []
for index, row in df.iterrows():
  sentences.append(row['Компани'] + ' ' + row['Салбар'])
s = sen_model.encode(sentences)
```

```python
#tsaling avaad float bolgon convert hiiged dundajiig n return hiine
def get_salary(salary_string):
    salary_string = salary_string.replace(',', '')
    if '-' in salary_string:
        low, high = salary_string.split('-')
        return (float(low) + float(high)) / 2
    else:
        return float(salary_string)

y = df['Цалин'].apply(get_salary) #tseverlesn tsalin column
reg = LinearRegression().fit(s, y)
```

```python
def predict_salary(query):
    query_embedding = sen_model.encode([query])
    prediction = reg.predict(query_embedding.reshape(1, -1))[0]
    return prediction

test1 = "Мобиком корпораци ХХК Мэдээлэл технологи, Харилцаа холбоо"
salary = predict_salary(test1)
formatted_salary = f"{salary:,.1f}₮"
print('Predicted Цалин: ', formatted_salary)
```
